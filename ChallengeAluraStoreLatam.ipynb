#1. Análisis de facturación



```python
# Facturación total por tienda
facturacion_1 = tienda["Precio"].sum()
facturacion_2 = tienda2["Precio"].sum()
facturacion_3 = tienda3["Precio"].sum()
facturacion_4 = tienda4["Precio"].sum()

print("Facturación Tienda 1:", facturacion_1)
print("Facturación Tienda 2:", facturacion_2)
print("Facturación Tienda 3:", facturacion_3)
print("Facturación Tienda 4:", facturacion_4)
```

```python
import matplotlib.pyplot as plt

```

####**Grafico de Barras**

```python
# Datos
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
facturacion = [facturacion_1, facturacion_2, facturacion_3, facturacion_4]

# Crear gráfico de barras
fig, ax = plt.subplots()
barras = ax.bar(tiendas, facturacion, color="lightgreen")

# Título y etiquetas
ax.set_title("Facturación total por tienda")
ax.set_ylabel("Facturación (en unidades monetarias)")
ax.set_xlabel("Tiendas")

# Mostrar los valores encima de cada barra
for i, valor in enumerate(facturacion):
    ax.text(i, valor + (valor * 0.01), f"{valor:,.0f}", ha="center", va="bottom", fontsize=9, color="black")

# Desactivar notación científica en el eje Y
ax.ticklabel_format(style="plain", axis="y")

plt.show()

```

# 2. Ventas por categoría

```python
def agrupar_categoria(dataset):
    return dataset.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
```

```python
# Ventas totales por categoría (Tienda 1)

tienda1_categorias = agrupar_categoria(tienda)
tienda1_categorias
```

```python

tienda2_categorias = agrupar_categoria(tienda2)
tienda2_categorias
```

```python

tienda3_categorias = agrupar_categoria(tienda3)
tienda3_categorias
```

```python
tienda4_categorias = agrupar_categoria(tienda4)
tienda4_categorias
```

###**Grafico de torta con la distribución de las ventas por categoría de producto**

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

tiendas = [
    (tienda1_categorias, 'Tienda 1'),
    (tienda2_categorias, 'Tienda 2'),
    (tienda3_categorias, 'Tienda 3'),
    (tienda4_categorias, 'Tienda 4')
]

for ax, (data, titulo) in zip(axes.flat, tiendas):
    wedges, texts, autotexts = ax.pie(
        data,
        labels=data.index,       # etiquetas de cada categoría
        autopct='%1.1f%%',       # porcentaje
        startangle=90,
        pctdistance=0.85,        # distancia del porcentaje al centro
        labeldistance=1.1,       # distancia de las etiquetas hacia afuera
        textprops={'fontsize': 9},
        colors=plt.cm.Paired.colors
    )
    ax.set_title(titulo, fontsize=12)
    ax.axis('equal')  # mantiene las tortas proporcionales

plt.suptitle('Distribución de ventas por categoría - Comparación entre tiendas', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Combinar los 4 DataFrames en uno solo
tiendas_combinadas = pd.concat([tienda, tienda2, tienda3, tienda4])

total_categorias = agrupar_categoria(tiendas_combinadas)
total_categorias

```

###**Grafico de tortas con la distribución total de ventas por categoría**

```python
fig, ax = plt.subplots(figsize=(8, 8))

wedges, texts, autotexts = ax.pie(
    total_categorias,
    labels=total_categorias.index,
    autopct='%1.1f%%',
    startangle=90,
    pctdistance=0.85,
    labeldistance=1.1,
    colors=plt.cm.Set3.colors,
    textprops={'fontsize': 10}
)

ax.set_title('Distribución total de ventas por categoría (todas las tiendas)', fontsize=14)
ax.axis('equal')

plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calcular el promedio de calificaciones de cada tienda
promedio_t1 = tienda['Calificación'].mean()
promedio_t2 = tienda2['Calificación'].mean()
promedio_t3 = tienda3['Calificación'].mean()
promedio_t4 = tienda4['Calificación'].mean()

print("Promedio de calificaciones por tienda:")
print(f"Tienda 1: {promedio_t1:.2f}")
print(f"Tienda 2: {promedio_t2:.2f}")
print(f"Tienda 3: {promedio_t3:.2f}")
print(f"Tienda 4: {promedio_t4:.2f}")
```

###**Grafico de barras con la calificacion promedio por tiendas**

```python
import matplotlib.pyplot as plt

# Crear un diccionario con los promedios
promedios = {
    'Tienda 1': promedio_t1,
    'Tienda 2': promedio_t2,
    'Tienda 3': promedio_t3,
    'Tienda 4': promedio_t4
}

# Crear el gráfico
plt.figure(figsize=(8,5))
barras = plt.bar(promedios.keys(), promedios.values(), color=plt.cm.Set2.colors)

# Título y etiquetas
plt.title('Calificación promedio por tienda')
plt.ylabel('Calificación promedio')
plt.ylim(0, 5)

# Mostrar valores encima de cada barra
for i, valor in enumerate(promedios.values()):
    plt.text(i, valor + 0.05, f'{valor:.2f}', ha='center', fontsize=10)

plt.show()
```

# 4. Productos más y menos vendidos

```python
# Contar cuántas veces se vendió cada producto en cada tienda
ventas_t1 = tienda['Producto'].value_counts()
ventas_t2 = tienda2['Producto'].value_counts()
ventas_t3 = tienda3['Producto'].value_counts()
ventas_t4 = tienda4['Producto'].value_counts()

# Mostrar el producto más y menos vendido de cada tienda
print("📈 Producto más vendido por tienda:")
print(f"Tienda 1: {ventas_t1.idxmax()} ({ventas_t1.max()} ventas)")
print(f"Tienda 2: {ventas_t2.idxmax()} ({ventas_t2.max()} ventas)")
print(f"Tienda 3: {ventas_t3.idxmax()} ({ventas_t3.max()} ventas)")
print(f"Tienda 4: {ventas_t4.idxmax()} ({ventas_t4.max()} ventas)")

print("\n📉 Producto menos vendido por tienda:")
print(f"Tienda 1: {ventas_t1.idxmin()} ({ventas_t1.min()} venta/s)")
print(f"Tienda 2: {ventas_t2.idxmin()} ({ventas_t2.min()} venta/s)")
print(f"Tienda 3: {ventas_t3.idxmin()} ({ventas_t3.min()} venta/s)")
print(f"Tienda 4: {ventas_t4.idxmin()} ({ventas_t4.min()} venta/s)")
```

###**Grafico de barras horizontal con el top 5 productos mas vendidos por tienda**

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
tiendas = [ventas_t1, ventas_t2, ventas_t3, ventas_t4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for i, (ventas, nombre) in enumerate(zip(tiendas, nombres)):
    fila = i // 2
    columna = i % 2

    top_5 = ventas.head(5)

    axes[fila, columna].barh(top_5.index, top_5.values, color='#4CAF50')
    axes[fila, columna].set_title(f'Top 5 productos más vendidos - {nombre}', fontsize=10, fontweight='bold')
    axes[fila, columna].set_xlabel('Cantidad vendida', fontsize=8)
    axes[fila, columna].set_ylabel('')
    axes[fila, columna].tick_params(axis='y', labelsize=8)
    axes[fila, columna].tick_params(axis='x', labelsize=8)
    axes[fila, columna].invert_yaxis()

    # Mostrar los valores al lado de cada barra, más chico
    for index, value in enumerate(top_5.values):
        axes[fila, columna].text(value + 0.3, index, str(value), va='center', fontsize=7)

plt.tight_layout(pad=2.0, w_pad=2.5, h_pad=2.5)
plt.show()
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio de cada tienda
promedio_envio_t1 = tienda['Costo de envío'].mean()
promedio_envio_t2 = tienda2['Costo de envío'].mean()
promedio_envio_t3 = tienda3['Costo de envío'].mean()
promedio_envio_t4 = tienda4['Costo de envío'].mean()

# Crear un diccionario con los resultados
promedios_envio = {
    'Tienda 1': promedio_envio_t1,
    'Tienda 2': promedio_envio_t2,
    'Tienda 3': promedio_envio_t3,
    'Tienda 4': promedio_envio_t4
}

```

###**Grafico de barras con costo promedio de envio por tienda**

```python
# Convertir a DataFrame para graficar
import pandas as pd
df_envio = pd.DataFrame(list(promedios_envio.items()), columns=['Tienda', 'Costo Promedio de Envío'])

# Crear gráfico de barras
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
barras = plt.bar(df_envio['Tienda'], df_envio['Costo Promedio de Envío'], color=['#1E88E5','#43A047','#FB8C00','#8E24AA'])
plt.title('Costo de Envío Promedio por Tienda', fontsize=13, fontweight='bold')
plt.ylabel('Costo promedio ($)', fontsize=10)
plt.xlabel('Tienda', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Mostrar el valor encima de cada barra
for i, valor in enumerate(df_envio['Costo Promedio de Envío']):
    plt.text(i, valor + 0.3, f'${valor:,.2f}', ha='center', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```
